```python
# Set seed (optional) ,,,,101 é um numero arbitrario
np.random.seed(101)

# Columns lembrar das colunas
print('ibovs')
print(ibov.columns)
print('\n')

# criar pesos aleatórios...ele cria n pesos diferentes, mas a soma deles não é 1
print('Creating Random Weights')
weights = np.array(np.random.random(73))
print(weights)
print('\n')

# Rebalanco dos pesos...agora a soma dos pesos é 1
print('Rebalance to sum to 1.0')
weights = weights / np.sum(weights)
print(weights)
print('\n')

# Retorno esperado...multiplica os retornos diario pelos pesos e multiplica por 365
print('Expected Portfolio Return')
exp_ret = np.sum(log_ret.mean() * weights) * days
print(exp_ret)
print('\n')

# Variancia experada (volatilidade) ...formula de algebra linear
print('Expected Volatility')
exp_vol = np.sqrt(np.dot(weights.T, np.dot(log_ret.cov() * days, weights)))
print(exp_vol)
print('\n')

# Sharpe Ratio
SR = exp_ret/exp_vol
print('Sharpe Ratio')
print(SR)
```

```python
#agora vamos pegar esse processo e repetir centenas de vezes para chegar a um resultado a respeito da alocação de portfolio
#quanto maior o numero de interações, mais tempo de processamento
```

```python
num_ports = 1000

#zerando os valores apos retornarem do loop
all_weights = np.zeros((num_ports,len(ibov.columns)))
ret_arr = np.zeros(num_ports)
vol_arr = np.zeros(num_ports)
sharpe_arr = np.zeros(num_ports)

for ind in range(num_ports): #maioria das equações abaixo paermancem, mas colocou um loop

    # Create Random Weights
    weights = np.array(np.random.random(73))

    # Rebalance Weights
    weights = weights / np.sum(weights)
    
    # Save Weights
    all_weights[ind,:] = weights

    # Expected Return
    ret_arr[ind] = np.sum((log_ret.mean() * weights) *days)

    # Expected Variance
    vol_arr[ind] = np.sqrt(np.dot(weights.T, np.dot(log_ret.cov() * days, weights)))

    # Sharpe Ratio
    sharpe_arr[ind] = ret_arr[ind]/vol_arr[ind]
```

```python
#melhor sharp ratio
sharpe_arr.max()
```

```python
#qual a posição dele...
sharpe_arr.argmax()
```

```python
#e qual a composição...
all_weights[sharpe_arr.argmax(),:]
```

```python
#vamos concatenar
securit=pd.DataFrame(ibov.columns)
allw=pd.DataFrame(all_weights[sharpe_arr.argmax(),:])*100
ibovret = pd.concat([securit,allw],axis=1)
ibovret_head = ['ticker', '% comp']
ibovret.columns=ibovret_head #colocando titulo nas colunas
pd.set_option('display.max_rows', None) # imprimindo a lista inteira
ibovret.sort_values(by='% comp', ascending=0) # imprimir em ordem descrecente
```
